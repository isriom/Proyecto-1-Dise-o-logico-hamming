```python
# Importación de Librerías para gráficar
import numpy as np
import matplotlib.pyplot as plt

def NRZI(number):
    # Señal viene en valor alto antes de t=0
    ultimoMovimientoImportante = "SUBIR"
    resultadoIntermedio = []
    for i in number:
        if i == "0":
            resultadoIntermedio += ["MANTENER"]
        else:
            if ultimoMovimientoImportante == "SUBIR":
                resultadoIntermedio += ["BAJAR"]
                ultimoMovimientoImportante = "BAJAR"
            else:
                resultadoIntermedio += ["SUBIR"]
                ultimoMovimientoImportante = "SUBIR"
    return resultadoIntermedio

def funcionNRZI(resultadoAConvertir):
    # Escalón unitario
    theFunction = "np.heaviside(t,1)"
    elementCounter = 0
    for element in resultadoAConvertir:
        if element == "SUBIR":
            theFunction +="+np.heaviside(t-"+str(elementCounter)+",1)"
        elif element == "BAJAR":
            theFunction +="-np.heaviside(t-"+str(elementCounter)+",1)"
        elementCounter += 1
    return theFunction

def plotNRZI(theFunction):
  u = lambda t: eval(theFunction)
  a = 0
  b = len(number)
  dt = 0.01

  # PROCEDIMIENTO
  ti  = np.arange(a, b, dt)
  u0_i = u(ti)

  plt.plot(ti,u0_i)

  plt.title('Señal NRZI')
  plt.ylabel('Amplitud')
  plt.xlabel('Tiempo (s)')
  plt.margins(dt)
  plt.show()

number = "100111100001"
plotNRZI(funcionNRZI(NRZI(number)))
number = "01011100"
plotNRZI(funcionNRZI(NRZI(number)))
```
